```python
import pandas as pd
import plotly.graph_objects as go
from datetime import datetime
import pytz
```

```python
df = pd.read_csv("google_play_store_dataset.csv")
df.columns = df.columns.str.strip()
df.head()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
df_filtered = df[
    (df['Reviews'] > 500) &
    (df['Category'].str.startswith(('E','C','B'), na=False)) &
    (~df['App'].str.startswith(('X','Y','Z'), na=False)) &
    (~df['App'].str.contains('S', case=False, na=False))
]
```

```python
translation = {
    "Beauty": "सौंदर्य",
    "Business": "வணிகம்",
    "Dating": "Dating"
}

df_filtered['Category'] = df_filtered['Category'].replace(translation)
```

```python
df_filtered['Month'] = df_filtered['Last Updated'].dt.to_period('M').dt.to_timestamp()
```

```python
time_data = df_filtered.groupby(['Month','Category'])['Reviews'].sum().reset_index()
```

```python
time_data['Growth'] = time_data.groupby('Category')['Reviews'].pct_change()
```

```python
fig = go.Figure()

categories = time_data['Category'].unique()

for cat in categories:
    data = time_data[time_data['Category'] == cat]

    fig.add_trace(
        go.Scatter(
            x=data['Month'],
            y=data['Reviews'],
            mode='lines+markers',
            name=cat,
            fill='tozeroy'
        )
    )
```

```python
growth_data = time_data[time_data['Growth'] > 0.20]

fig.add_trace(
    go.Scatter(
        x=growth_data['Month'],
        y=growth_data['Reviews'],
        mode='markers',
        marker=dict(size=10, symbol='circle'),
        name="Significant Growth"
    )
)
```

```python
fig.update_layout(
    title="Trend of Total Installs Over Time by App Category",
    xaxis_title="Time",
    yaxis_title="Total Installs (Reviews)",
)
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)

if 18 <= current_time.hour < 21:
    fig.show()
else:
    print("Graph only visible between 6 PM and 9 PM IST")
```
