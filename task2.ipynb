```python
df = pd.read_csv("google_play_store_dataset.csv")

df.head()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df_filtered = df[~df['Category'].str.startswith(('A','C','G','S'), na=False)]
```

```python
top_categories = df_filtered.groupby('Category')['Reviews'].sum().nlargest(5).index

df_top = df_filtered[df_filtered['Category'].isin(top_categories)]
```

```python
map_data = df_top.groupby(['Country','Category'])['Reviews'].sum().reset_index()
```

```python
map_data['Highlight'] = map_data['Reviews'] > 1000000
```

```python
fig = px.choropleth(
    map_data,
    locations="Country",
    locationmode="country names",
    color="Reviews",
    hover_name="Category",
    title="Global Installs by App Category",
    color_continuous_scale="Viridis"
)
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)

if 18 <= current_time.hour < 20:
    fig.show()
else:
    print("Graph only visible between 6 PM and 8 PM IST")
```
