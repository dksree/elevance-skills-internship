```python
df = pd.read_csv("googleplaystore.csv")

df.head()
```

```python
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')

df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

```python
df_filtered = df[~df['Category'].str.startswith(('A','C','G','S'))]

df_filtered.head()
```

```python
top_categories = df_filtered.groupby('Category')['Installs'].sum().nlargest(5)

top_categories
```

```python
df_top = df_filtered[df_filtered['Category'].isin(top_categories.index)]

df_top.head()
```

```python
category_data = df_top.groupby('Category')['Installs'].sum().reset_index()

category_data
```

```python
category_data['Highlight'] = category_data['Installs'] > 1000000

category_data
```

```python
category_data['Country'] = ['USA','IND','BRA','CAN','AUS'][:len(category_data)]

category_data
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)

print("Current IST Time:", current_time)
```

```python
if 18 <= current_time.hour < 20:

    fig = px.choropleth(
        category_data,
        locations="Country",
        locationmode="ISO-3",
        color="Installs",
        hover_name="Category",
        title="Top 5 App Categories by Installs",
        color_continuous_scale="Plasma"
    )

    fig.show()

else:
    print("Graph will only appear between 6 PM and 8 PM IST")
```
