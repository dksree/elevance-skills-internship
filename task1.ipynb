```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import pytz
import plotly.express as px
```

```python
df = pd.read_csv("google_play_store_dataset.csv")
df.head()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
```

```python
df['Size'] = df['Size'].astype(str).str.replace('M','')
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
df_filtered = df[
    (df['Size'] > 10) &
    (df['Last Updated'].dt.month == 1)
]
```

```python
top_categories = df_filtered.groupby('Category')['Reviews'].sum().nlargest(10).index

df_top = df_filtered[df_filtered['Category'].isin(top_categories)]
```

```python
summary = df_top.groupby('Category').agg({
    'Rating':'mean',
    'Reviews':'sum'
}).reset_index()
```

```python
summary = summary[summary['Rating'] >= 4.0]
```

```python
fig = px.bar(
    summary,
    x='Category',
    y=['Rating','Reviews'],
    barmode='group',
    title="Average Rating vs Total Reviews for Top App Categories"
)
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)

if 15 <= current_time.hour < 17:
    fig.show()
else:
    print("Graph only visible between 3 PM and 5 PM IST")
```
