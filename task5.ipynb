```python
import pandas as pd
import plotly.express as px
from datetime import datetime
import pytz
```

```python
df = pd.read_csv("google_play_store_dataset.csv")

df.columns = df.columns.str.strip()

df.head()
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

df['Size'] = df['Size'].astype(str).str.replace('M','')
df['Size'] = pd.to_numeric(df['Size'], errors='coerce')
```

```python
categories = [
    "GAME","BEAUTY","BUSINESS","COMICS","COMMUNICATION",
    "DATING","ENTERTAINMENT","SOCIAL","EVENTS"
]

df_filtered = df[df['Category'].str.upper().isin(categories)]
```

```python
df_filtered = df_filtered[
    (df_filtered['Rating'] > 3.5) &
    (df_filtered['Reviews'] > 500) &
    (df_filtered['Reviews'] > 50000) &
    (~df_filtered['App'].str.contains('S', case=False, na=False))
]
```

```python
translations = {
    "BEAUTY": "सौंदर्य",
    "BUSINESS": "வணிகம்",
    "DATING": "Dating"
}

df_filtered['Category'] = df_filtered['Category'].str.upper().replace(translations)
```

```python
color_map = {
    "GAME": "pink"
}

fig = px.scatter(
    df_filtered,
    x="Size",
    y="Rating",
    size="Reviews",
    color="Category",
    hover_name="App",
    title="App Size vs Rating with Installs",
)
```

```python
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist)

if 17 <= current_time.hour < 19:
    fig.show()
else:
    print("Graph only visible between 5 PM and 7 PM IST")
```
